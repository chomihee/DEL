# ETL 함수를 하나씩 정의

```
import requests

def extract(url):
    f = requests.get(url)
    return (f.text)
```

```
def transform(text):
    lines = text.split("\n")
    return lines
```

```
# 과제 수정영역

def load(lines):
    # BEGIN과 END를 사용해서 SQL 결과를 트랜잭션으로 만들어주는 것이 좋음
    # BEGIN;DELETE FROM (본인의스키마).name_gender;INSERT INTO TABLE VALUES ('KEEYONG', 'MALE');....;END;
    cur = get_Redshift_connection()
    cur.execute("DELETE FROM good422.name_gender")
    for r in lines[1:]:
        if r != '':
            (name, gender) = r.split(",")
            print(name, "-", gender)
            sql = "BEGIN ;\
                    INSERT INTO good422.name_gender VALUES ('{n}', '{g}'); \
                   END; ".format(n=name, g=gender)
            print(sql)
            cur.execute(sql)
```
